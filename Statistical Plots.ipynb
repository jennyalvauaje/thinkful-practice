```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generating 1000 random values
x = np.random.normal(10, 5, 1000)

#plot as histogram
plt.hist(x)
plt.show()

#10 & 5 are the center and stddev of the normal curve


```

```python
x = np.random.normal(10, 5, 1000)

plt.hist(x, bins=80, color='red')
plt.title('Default Bin Placement Demo')
plt.xlabel('Random Values')
plt.show()
```

```python
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

plt.hist(x, color='blue', bins=np.arange(-10, 40), alpha = 0.5)
plt.hist(y, color='red', bins=np.arange(-10, 40), alpha = 0.5)

plt.title('Manually setting bin placement')
plt.xlabel('Random Values')

plt.show()
```

```python
#normalize this data for comparison
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

plt.hist(x, density=True, color='blue', bins=np.arange(-10, 40), alpha = 0.5)
plt.hist(y, density=True, color='red', bins=np.arange(-10, 40), alpha = 0.5)

plt.title('Normed Histograms')
plt.xlabel('Random Values')

plt.show()
```

```python
#boxplot

np.random.seed(111)

x = np.random.normal(10, 5, 1000)
plt.boxplot(x)
plt.show()
```
