```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as stats

plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
football = pd.read_csv("/Users/Jenny/Documents/Thinkful/international football data/results.csv")
football.head()
```

```python
score_away = sns.factorplot(x="home_score", y="away_score", data=football,
                   size=6, kind="bar", palette="pastel", ci=95)
score_away.despine(left=True)
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.title('Barplot: Away Scores by Home Scores')
plt.xlim(-1, 16.5)
plt.show()
```

```python
sns.plt.rc("figure", figsize=(20.0,10.0))
score_pointplot = sns.factorplot(x="home_score", y="away_score", data=football,
                   kind="point", palette="muted")
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.title('Pointplot: Away Scores by Home Scores')
plt.show()
```

```python
score_boxplot = football.boxplot(column = 'home_score', by = 'away_score')
plt.title('Boxplot: Away Scores by Home Scores')
plt.suptitle("")
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.ylim(0,7)
plt.show()
```

```python
plt.scatter(x=football['home_score'], y=football['away_score'])
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.title('Scatterplot: Away Scores by Home Scores')
plt.show()
```

```python
sns.stripplot(x=football['home_team'], y=football['home_score'], data=football, jitter = True)
plt.xlim(-1, 19.5)
plt.xticks(rotation=90)
plt.xlabel('Home Team')
plt.ylabel('Home Score')
plt.title('Home Scores by Home Team')
```

```python
sns.stripplot(x="home_team", y="home_score", hue = "neutral", data=football)
plt.xlim(-1, 19.5)
plt.xticks(rotation=90);
```

```python
sns.violinplot(x="home_team", y="home_score", hue="neutral", data=football, split = True)
plt.xlim(-1,19.5)
plt.xticks(rotation=90);
```

```python
plt.subplot(1, 2, 1)
plt.ylim(0,10000)
sns.distplot(football.away_score, kde=False)
plt.xlabel('Away Score')
plt.ylabel('Frequency')
plt.title('Away Score Frequency Distribution')

plt.subplot(1,2,2)
plt.ylim(0,10000)
sns.distplot(football.away_score, kde=False, rug = True)
plt.xlabel('Away Score')
plt.ylabel('Frequency')
plt.title('Away Score Frequency Distribution')
```

```python
sns.distplot(football.away_score, fit = stats.gamma)
plt.xlabel('Away Score')
plt.ylabel('Density')
plt.title('Away Score Frequency/KDE')
```
